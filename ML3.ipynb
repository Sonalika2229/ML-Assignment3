a. Using NumPy create random vector of size 15 having only Integers in the range 1-20.
1. Reshape the array to 3 by 5
2. Print array shape.
3. Replace the max in each row by 0
Create a 2-dimensional array of size 4 x 3 (composed of 4-byte integer elements), also print the shape, type and data type
of the array

```python
import numpy as np

# Creating a random vector of size 15 with integers of range 1-20
a = np.random.randint(low=1, high=21, size=15)

# Reshaping the vector to 3 by 5 array
a_3x5 = a.reshape(3, 5)

# Printing the shape of the array
print(a_3x5.shape)

# Replacing the max in each row by 0
a_3x5[np.arange(3), a_3x5.argmax(axis=1)] = 0

# Print the modified array
print(a_3x5)
```

# b.Write a program to compute the eigenvalues and right eigenvectors of a given square array given below:
[[ 3 -2]
[ 1 0]]

```python
# importing numpy library
import numpy as np
  
# create numpy 2d-array
n = np.array([[3,-2,],
              [1,0]])
# finding eigenvalues and eigenvectors
x, y = np.linalg.eig(n)
  
# printing eigen values
print("Eigen values of the given square array:\n", x)
  
# printing eigen vectors
print("Right eigenvectors of the given square array:\n", y)
```

c.Compute the sum of the diagonal element of a given array.
[[0 1 2]
[3 4 5]]

```python
import numpy as np

# Define the array
a = np.array([[0, 1, 2], [3, 4, 5]])

# Compute the sum of the diagonal elements
sum_of_diagonal_elements = np.trace(a)

# Print the sum of the diagonal elements
print("Sum of the diagonal elements:", sum_of_diagonal_elements)
```

d. Write a NumPy program to create a new shape to an array without changing its data.
Reshape 3x2:
[[1 2]
[3 4]
[5 6]]
Reshape 2x3:
[[1 2 3]
[4 5 6]]

```python
import numpy as np

# Defining the original array
array = np.array([[1, 2], [3, 4], [5, 6]])

# Reshape the array to 2x3
arr1 = np.reshape(array, (2, 3))

# Print the original and new arrays
print("original array:")
print(array)
print("Reshape 2x3:")
print(arr1)
```

Matplotlib
1. Write a Python programming to create a below chart of the popularity of programming Languages.
2. Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt
# Data to plot
lang = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
Popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
Colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467be", "#8c564b"]
# explode 1st slice
explode = (0.1, 0, 0, 0,0,0)  
# Plot
plt.pie(Popuratity, explode=explode, labels=lang, colors=Colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
